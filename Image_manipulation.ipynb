```python
import matplotlib.pyplot as plt
from skimage import io
from skimage import color
from skimage.transform import rotate
from skimage.transform import resize
```

```python
image = io.imread('/content/rose.jpg')
```

```python
io.imshow(image)
```

```python
im2= io.imsave('image2.jpg',image)
```

```python
im3 = color.rgb2hsv(image)
```

```python
img4 = resize(im3,(40,40))
img5= rotate(img4,(50))
```

```python
io.imshow(img5)
```
