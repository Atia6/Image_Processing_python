```python

from skimage import io
import cv2
```

```python
image =  io.imread('/content/rose.jpg')
```

```python
image2 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
image3 = cv2.medianBlur(image,3)
```

```python
image4 = cv2.equalizeHist(image2)
```

```python
io.imshow(image2)
```

```python
io.imshow(image3)
```

```python
io.imshow(image4)

```

```python
io.imsave('image_a.jpg',image2)
io.imsave('image_b.jpg',image3)
io.imsave('image_c.jpg',image4)
```
